4. Introducing Marketing Campaigns
----------------------------------

Marketing campaigns can influence the rates of acquiring new customers and retaining existing ones.


Let's simulate the impact of a marketing campaign that increases the rate of acquiring new customers $r$ for a certain period.


```python
# Parameters
B0 = 100  # Initial brand size (initial number of customers)
r_initial = 0.05  # Initial rate of acquiring new customers
r_campaign = 0.1  # Rate during marketing campaign
s = 0.03  # Rate at which customers stop buying the product
t = np.linspace(0, 50, 400)  # Time range

# Brand size with and without campaign
B_no_campaign = B0 * np.exp((r_initial - s) * t)

# During campaign (assume campaign lasts from t=10 to t=20)
t_campaign = np.linspace(10, 20, 100)
B_campaign = B0 * np.exp((r_initial - s) * t)
B_campaign[t > 10] = B_campaign[t <= 10][-1] * np.exp((r_campaign - s) * (t[t > 10] - 10))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, B_no_campaign, label='Without Campaign')
plt.plot(t[t > 10], B_campaign[t > 10], label='During Campaign', linestyle='--')
plt.title('Impact of Marketing Campaign on Brand Size')
plt.xlabel('Time')
plt.ylabel('Brand Size (Number of Customers)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Parameters
B0 = 100  # Initial brand size (initial number of customers)
r = 0.05  # Rate of acquiring new customers
s_values = [0.01, 0.03, 0.05]  # Different retention rates
t = np.linspace(0, 50, 400)  # Time range

# Plotting different retention rates
plt.figure(figsize=(10, 6))
for s in s_values:
    B = B0 * np.exp((r - s) * t)
    plt.plot(t, B, label=f'Retention rate = {s}')

plt.title('Impact of Retention Rate on Brand Size')
plt.xlabel('Time[weeks]')
plt.ylabel('Brand Size (Number of customers)')
plt.legend()
plt.grid(True)
plt.show()
```
