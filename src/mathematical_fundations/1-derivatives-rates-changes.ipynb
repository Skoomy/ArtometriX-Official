
## Derivatives

Calculus is the science of measuring change. Early in its history, its tools were developed to solve problems involving the position, velocity, and acceleration of moving objects. 


Prior to the development of calculus, there was no way to express this change in a variable. In this section, we introduce the notion of limits to develop the derivative of a function. 


The derivative, commonly denoted as f'(x), will measure the instantaneous rate of change of a function at a certain point x = a.

This number f'(a), when defined, will be graphically represented as the slope of the tangent line to a curve. 


****


- The average rate of change of the function $f(x)$ over the interval x=a to x=b is 
$$\frac{\Delta f}{\Delta x} = \frac{f(b) - f(a)}{b-a} $$

â€‹
- The instantaneous rate of change of a function x=c is the limit of the average rates of change of f over shorter and shorter intervals containing c.

- The instantaneous rate of change of $f$ at x=c is also called the derivative, and denoted by $f'(c)$.


The limit definition of the derivative states that 
$$ f'(c) = \lim_{h \to 0} \frac{f(c+h) - f(c)}{h}. $$

$[c, c+h]$ is an interval containing $(c)$. 

$ \frac{f(c+h) - f(c)}{(c+h) - c} = \frac{f(c+h) - f(c)}{h}$

is the average rate of change of \(f\) over this interval. 
As \(h\) goes to \(0\), the interval gets shorter and shorter.


***

- Average rate of change over an interval = slope of a secant line between two points.
  
- Instantaneous rate of change at a value = slope of a tangent line at a point.



The tangent line to the graph of $f(x)$ at the point $x = a$ has the equation $y = f(a) + f'(a)(x - a)$.




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
 # Define the function f(x)
def f(x):
    return x**2

# Derivative of f(x)
def df(x):
    return 2*x

# Points for the function plot
x = np.linspace(-2, 3, 400)
y = f(x)



# Points for the secant and tangent lines
c = 1
h = 1

x_secant = np.array([c, c+h])
y_secant = f(x_secant)
```

```python
# Tangent line at c
x_tangent = np.linspace(c-1, c+1, 10)
y_tangent = df(c) * (x_tangent - c) + f(c)
```

```python
plt.figure(figsize=(10, 6))

# Plot the function f(x)
plt.plot(x, y, label='$f(x) = x^2$')


# Plot the secant line
plt.plot(x_secant, y_secant, 'bo-', label='Secant Line')

# Plot the tangent line
plt.plot(x_tangent, y_tangent, 'r--', label='Tangent Line')

# Mark points P and Q
plt.plot(c, f(c), 'ro')  # Point P
plt.text(c, f(c), f' P({c}, {f(c)})', fontsize=12, verticalalignment='bottom')
plt.plot(c+h, f(c+h), 'ro')  # Point Q
plt.text(c+h, f(c+h), f' Q({c+h}, {f(c+h)})', fontsize=12, verticalalignment='bottom')

# Labels and legend
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Visualizing Rates of Change')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.show()
```

```python
h_values = np.linspace(0.001,0.1,5)
plt.figure(figsize=(10, 6))

plt.plot(x, y, label='$f(x) = x^2$')

for h in h_values:
    x_secant = np.array([c, c+h])
    y_secant = f(x_secant)
    plt.plot(x_secant, y_secant, 'bo-', alpha=0.6)

# Plot the tangent line
plt.plot(x_tangent, y_tangent, 'r--', label='Tangent Line')


# Mark point P
plt.plot(c, f(c), 'ro')  # Point P
plt.text(c, f(c), f' P({c}, {f(c)})', fontsize=12, verticalalignment='bottom')

# Labels and legend
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Visualizing Rates of Change with Multiple Secant Lines')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, diff, lambdify

# Define the symbolic variable and function
x = symbols('x')
f = 4*x**4 - x**3 - 6*x**2 + x + 1

# Calculate the derivative
df = diff(f, x)

# Define the function for the tangent line
def tangent_line_f(a):
    f_a = f.subs(x, a)
    m = df.subs(x, a)
    return f_a + m*(x - a)

# Define the function to plot using lambdify
f_np = lambdify(x, f, 'numpy')
df_np = lambdify(x, df, 'numpy')

# Define the x values for plotting
x_vals = np.linspace(-2, 2, 400)
y_vals = f_np(x_vals)

plt.figure(figsize=(10, 6))

# Plot the function f(x)
plt.plot(x_vals, y_vals, label='$f(x) = 4x^4 - x^3 - 6x^2 + x + 1$')

# Plot tangent lines for different values of a
for k in [1, 2, 3, 4, 5]:
    tangent_line_expr = tangent_line_f(k)
    tangent_line_np = lambdify(x, tangent_line_expr, 'numpy')
    y_tangent_vals = tangent_line_np(x_vals)
    plt.plot(x_vals, y_tangent_vals, '--', label=f'Tangent line at x={k}')

# Labels and legend
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Function and Tangent Lines')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.show()
```
