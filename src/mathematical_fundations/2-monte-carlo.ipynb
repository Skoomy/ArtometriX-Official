Probability to find random location exploration at location x and time t is given by :
\begin{equation}
    \begin{aligned}
        p(t,x) &= \sum_{x'} p(t,x') W_{x'\rightarrow x}\\
         & = \sum_{x'\neq x} p(t,x') W_{x'\rightarrow x} + p(t,x) W_{x\rightarrow x} \\
         & = \sum_{x'\neq x} p(t,x') W_{x'\rightarrow x} + p(t,x) (1 - \sum_{x' \neq x} W_{x\rightarrow x'}) \\
        & = p(t,x) + \sum_{x' \neq x} [p(t,x') W_{x\rightarrow x'} - p(t,x) W_{x\rightarrow x'})] \\
    \end{aligned}
\end{equation}

```python
import numpy as np

class Integrator:
    def __init__(self, xMin, xMax, N):
        self.xMin = xMin
        self.xMax = xMax
        self.N = N
        self.dx = (xMax - xMin) / N
        self.result = None

    def f(self, x):
        return x**2 * np.exp(-x) * np.sin(x)

    def integrate(self):
        x_values = np.linspace(self.xMin, self.xMax, self.N, endpoint=False)
        f_values = self.f(x_values)
        self.result = np.sum(f_values) * self.dx

    def show(self):
        if self.result is not None:
            print(f"The result of the integration is: {self.result}")
        else:
            print("You need to perform the integration first using the integrate method.")
```

```python
integrator = Integrator(xMin=1, xMax=3,N= 200000)
integrator.integrate()
integrator.show()
```

## Continuous dynamical system Growth of a population

$P(t)$: number of individuals of a population

$P(0) = P_{0}$
Only two process are affecting the process: 

- the number of death $D(T)$
- the number of birth $B(T)$



> A possible model for Brith 

$$B(t) = r_{b} * P(t)$$

> A Possible model for Death:
> 
$$D(t) = r_{d} * P(t)$$

Variation of population can be expressed as 

$$P'(t) = B(t) - D(t) = (r_b - r_t) P(t) = r P(t)$$
The solution of this equation is 

$$P(t)= P_0 \exp(rt)$$

remember it's a rate so 0<r<1

**Infinite growth can not happend**
We can have a target Capacity $C$ to the population as $r \rightarrow 0$ as $P\rightarrow C$



$$r = \nu (1 - \frac{P}{C})$$

$$\frac{dP(t)}{dt} = \nu (1 - \frac{P}{C}) P $$

solution is 

$$P(t) = \frac{C}{1+\frac{C-P_0}{P_0} exp(-\nu t )}$$
P_0 < C growth and P -> C 
p_0 > C decay of population and P-> C
Except if P_0 = 0 then P(t) = 0

Two fixed points C (attractive) and 0 repulsive

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
P0 = 10  # Initial population
r_growth = 0.05   # Growth rate
r_decay = -0.1
t = np.linspace(0, 10, 40)  # Time range

# Exponential growth formula
p_g = P0 * np.exp(r_growth  * t)
p_d = P0 * np.exp(r_decay  * t)
p_total = p_g - p_d
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, p_g, label='Exponential Growth',color='blue')
plt.plot(t, p_total, label='Total population',color='orange')

plt.axhline(P0, label='Constant',color='black',ls='--')
plt.plot(t, p_d, label='Exponential Decay',color='red')
plt.title('Exponential Growth of Population')
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.legend()
plt.grid(True)
plt.show()
```

### Lotka- Volterra model (Prey-Predator)

Also known as predator-prey equations, describe the variation in populations of two species which interact via predation. 


- a be the antilope(Prey) population
- c the population of cheetah (Predator)

<div class="alert alert-block" style="border-color:black balck black black;background-color:pink">
<ul>
    <li>The antelopes reproduce exponentially fast.</li> 
    <li>The antelopes are being eaten by the cheetah proportionally to their chance to meet.</li>
    <li>The cheetahs die exponentially fast of starvation.</li>
    <li>The cheetahs are reproducing proportionally to their chance to eat antelopes.</li>
</ul>
</div>



$$
\frac{da}{dt} = k_a a(t) - k_{c,a} c(t) a(t)
$$

$$
\frac{dc}{dt} = - k_{c} c(t) + k_{a,c} a(t) c(t)
$$

- $k_{c,a}c(t)a(t)$ signifies the death rate of the antelope
- $k_{a}a(t)$ signifies the reproduction rate of the antelope.
- $\frac{da}{dt}$ signifies the variation of the antelope population.

- So here we assume that the cheetahs do not reproduce on their own if they are not fed. Which means that in absence of any antelope, the cheetahs will simply starve and die.

-  $k_{a,c} a(t) c(t)$ the growth of the population, which is proportional to the the probabilities that cheetah eat antelopes

```python
from scipy import integrate
```

```python



alpha = 1. #mortality rate due to predators
beta = 1.
delta = 1.
gamma = 1.

# intial condition 
a_0 = 3. #for antiloppe 
c_0 = 4. #for chetat

def derivative(X, t, alpha, beta, delta, gamma):
    x, y = X
    dotx = x * (alpha - beta * y)
    doty = y * (-delta + gamma * x)
    return np.array([dotx, doty])


```

```python
Nt = 1000
tmax = 30.
t = np.linspace(0.,tmax, Nt)
X0 = [a_0, c_0]
res = integrate.odeint(derivative, 
                       X0, 
                       t, 
                       args = (alpha, beta, delta, gamma))
x, y = res.T
```

```python
plt.figure()
plt.grid()
# Time evolution of the two population
plt.title("Lotka-Volterra model\n Time evolution Periodicity")
plt.plot(t, x, 'xb', label = 'Antilope')
plt.plot(t, y, '+r', label = "Cheetah")
plt.xlabel('Time t, [days]')
plt.ylabel('Population a(t), c(t)')
plt.legend()

plt.show()
```

>  So what we see here is that basically for any initial condition, we'll be on a loop on this parametric plot. So we see that when we have a low number of cheetahs, we'll usually have relatively large amount of antelopes and vice versa

```python
plt.figure()
IC = np.linspace(1.0, 6.0, 10) # initial conditions for deer population (prey)
for deer in IC:
    X0 = [deer, 1.0]
    Xs = integrate.odeint(derivative, X0, t, args = (alpha, beta, delta, gamma))
    plt.plot(Xs[:,0], Xs[:,1], "-", label = "$x_0 =$"+str(X0[0]))
plt.xlabel("Antilope")
plt.ylabel("Cheetah")
plt.legend()
plt.title("Antilopes vs Cheetah");
```
