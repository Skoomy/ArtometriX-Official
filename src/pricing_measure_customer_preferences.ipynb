## Willingness to pay from survey. 

`keywords`: Pricing strategy 

Context as a customer you are looking to buy some new carbon wheel for your bike. You found a pair of wheel at a price of 
150 , first reaction would be this is to cheap for some carbon wheel they must be of really poor quality.


As pricing analyst you are task with goal of setting the price of a new pair of carbon wheel the idea is to conduct a survey and ask people how much they are willing to pay for some good carbon wheel.



First step Data collection:

Survey no.1: At what price the carbon wheel would be so inexpensive that you would doubt it's quality. Now, what is that mean?



```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

```

### Sample response of minimum price 

> Represents accetable low price.

```python
np.random.seed(7)
range_price = np.linspace(400,2000,9)
result_survey = np.random.choice(range_price,3_000)

data = pd.DataFrame({'minimum_price':range_price})

# data['frequency']

data = np.unique(result_survey, return_counts=True)
data = pd.DataFrame(np.asarray(data).T,columns=['minimum_price','frequency'])
data['percentage_pct']=data['frequency'] *100 / data['frequency'].sum()
data=data.sort_values(by=['minimum_price'])
data['cumsum_pct'] = data['percentage_pct'].cumsum()

print("\nSample response of minimum price ie represents accetable low price.\n\n")

display(data)
```

**Survey results**

> Let's look at 1400.0	 and that's about 69.95%. What does that mean? It means for 69.95% of your target population 1400.0 is an acceptably low price.




### Sample response at maximum price

> Price at which product not worth buying

```python
np.random.seed(7)

range_price = np.linspace(800,4000,17)
result_survey = np.random.choice(range_price,3_000)


# data['frequency']

data_max = np.unique(result_survey, return_counts=True)
data_max = pd.DataFrame(np.asarray(data_max).T,columns=['max_price','frequency_max'])
data_max['max_price_percentage_pct']=data_max['frequency_max'] *100 / data_max['frequency_max'].sum()
data_max=data_max.sort_values(by=['max_price'])
data_max['cumsum_pct_max'] = data_max['max_price_percentage_pct'].cumsum()


print("Sample response at which product not worth buying")

display(data_max)
```

### Combined survey data

this data gives you : 

-  The percentage of people who feel the price is not too low.
-  The percentage of people who feel the price is to high

Based on this data we can get peoples willingness to pay and the number of people that would pay that



```python
datum = []
for e in data.to_dict(orient='records'):
    res = {}
    res['price'] = e['minimum_price']
    # res['frequency'] = e['frequency']
    # res['min_price_percentage_pct'] = e['percentage_pct']
    # res['max_price_percentage_pct'] =  max_price_res['max_price_percentage_pct']
    res['min_price_cumsum'] = e['cumsum_pct']
    if e['minimum_price'] in data_max['max_price'].unique():
        max_price_res = data_max[data_max['max_price'] == e['minimum_price']].to_dict(orient='records')[0]
        res['max_price_cumsum'] = max_price_res['cumsum_pct_max']
        
        
       
    else:
        # res['max_price_percentage_pct'] = 0
        res['max_price_cumsum'] = 0
    
    res['willing_to_pay'] = np.absolute(res['min_price_cumsum'] - res['max_price_cumsum'])

    datum.append(res)



for e in data_max.to_dict(orient='records'):
    if e['max_price'] not in data['minimum_price'].unique():
        res = {}
        res['price'] = e['max_price']
        # res['frequency'] = e['frequency_max']
        # res['min_price_percentage_pct'] = 0
        # res['max_price_percentage_pct'] =  max_price_res['max_price_percentage_pct']
        res['max_price_cumsum'] = e['cumsum_pct_max']
        res['min_price_cumsum' ]= 100
        res['willing_to_pay'] = res['min_price_cumsum'] - res['max_price_cumsum']

        datum.append(res)
    
datum = pd.DataFrame(datum).round()
datum
```

#### Interpreting results

> This means that 33% of the target population finds 1000 EUR to be an acceptable price: it's not so low that they would consider the quality poor, nor too high to deter them from purchasing if they otherwise like the product.


### Factor in cost 

The previous data is incomplete as it doesn't account for the production cost of the wheel. Therefore, we need to incorporate some business financials:



-  Assume the cost of producing the wheel is 700 EUR.
  
-  Margin is calculated by subtracting the cost from the selling price.
  $$margin = price - cost$$

- Profit is determined by multiplying the margin by the cumulative percentage of people willing to pay that specific price.
  $$ profit = margin * cumsum_{rightPrice}$$


```python
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow; color: red;' if v else '' for v in is_max]



datum['cost'] = 700
datum['margin'] = (datum['price'] - datum['cost'])
datum['profit'] = datum['margin'] * datum['willing_to_pay']/100
styled_datum = datum.style.apply(highlight_max, subset=['profit'])
styled_datum
```

### Results

Based on the survey data and our indirect approach to assessing willingness to pay, the data suggests that the optimal price, given a cost of 700 EUR, would be 2400 EUR. This method proves more effective than directly asking people how much they’d be willing to pay.

## Willingness to pay from conjoint analysis 


```python
# Define attributes and levels
attributes = {
    "capacity_oz": [18, 24, 30, 36],
    "cost_dollar": [25, 32, 45, 50],
    "cold_retention_h": [6, 12, 18, 24],
    "product_name": [f'product_{i}' for i in range(1,5)]
}

# Create a list of all possible product profiles
product_profiles = pd.DataFrame(attributes
)


product_profiles
```

###  Collect preference data 

- synthethic data : Generate integer preference data on a scale of 1 to 5 for 100 respondents


```python
num_respondents = 100

preferences = pd.DataFrame(
    np.random.randint(1, 6, size=(num_respondents, len(product_profiles))),
    columns=product_profiles['product_name'].unique()
)
display(preferences)
```

```python
# Reshape the preferences data for analysis
preferences_long = preferences.melt(var_name='profile', value_name='preference')
preferences_long = preferences_long.sort_values(by='profile').reset_index(drop=True)

# Merge the preferences with the product profiles
preferences_long = preferences_long.merge(
    product_profiles, left_on='profile', 
    right_on='product_name'
)

display(preferences_long)
```

```python
# One-hot encode the attributes
preferences_encoded = pd.get_dummies(preferences_long.drop(columns=['profile','product_name']))
preferences_encoded
```

### Estimates utility values 

> To find the increased willingness to pay for extending the heat retention time from 12 hours to 18 hours, we calculate the utility difference and convert it into monetary terms:

```python

# Fit a linear regression model to estimate utility values
X = preferences_encoded.drop(columns=['preference'])
y = preferences_encoded['preference']
model = LinearRegression()
model.fit(X, y)

# Extract the utility values
utility_values = pd.Series(model.coef_, index=X.columns)
utility_values

# Print utility values for each attribute level
print("Utility Values:")
print(utility_values)

# # Summarize the utility values by attribute
# utility_summary = utility_values.groupby(lambda x: x.split('_')[0]).mean()
# print("\nUtility summary:")
# print(utility_summary)
```
